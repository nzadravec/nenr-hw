```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
f = open("samples.txt", "r")
lines = f.readlines()
f.close()

z = np.empty(len(lines))
z1 = np.empty(len(lines))
x = np.empty(len(lines)) 
y = np.empty(len(lines))

for i, line in enumerate(lines):
    a, b, c, d = line.split(" ")
    y[i] = float(a)
    x[i] = float(b)
    z[i] = float(c)
    z1[i] = float(d)
    
X = x.reshape((80,80))
Y = y.reshape((80,80))
Z = z.reshape((80,80))
Z1 = z1.reshape((80,80))

fig = plt.figure(figsize=(12,24))

ax = fig.add_subplot(4,2,2, projection='3d')
ax.plot_surface(X, Y, Z, rstride=4, cstride=4)
ax.view_init(30, 45)

ax = fig.add_subplot(4,2,1, projection='3d')
ax.plot_surface(X, Y, Z1, rstride=4, cstride=4)
ax.view_init(30, 45)

ax = fig.add_subplot(4,2,4, projection='3d')
ax.plot_surface(X, Y, Z, rstride=4, cstride=4)
ax.view_init(70, 30)

ax = fig.add_subplot(4,2,3, projection='3d')
ax.plot_surface(X, Y, Z1, rstride=4, cstride=4)
ax.view_init(70, 30)

ax = fig.add_subplot(4,2,6, projection='3d')
ax.plot_surface(X, Y, Z, rstride=4, cstride=4)
ax.view_init(0,0)

ax = fig.add_subplot(4,2,5, projection='3d')
ax.plot_surface(X, Y, Z1, rstride=4, cstride=4)
ax.view_init(0,0)

ax = fig.add_subplot(4,2,8, projection='3d')
ax.plot_surface(X, Y, Z, rstride=4, cstride=4)
ax.view_init(0,90)

ax = fig.add_subplot(4,2,7, projection='3d')
ax.plot_surface(X, Y, Z1, rstride=4, cstride=4)
ax.view_init(0,90)

fig.tight_layout()
fig.savefig("fukcije.pdf")
```

```python
fig = plt.figure(figsize=(12,12))
zlim = 1

ax = fig.add_subplot(2,2,1, projection='3d')
ax.plot_surface(X, Y, Z-Z1, rstride=4, cstride=4)
ax.view_init(30, 45)
ax.set_zlim(-zlim, zlim)

ax = fig.add_subplot(2,2,2, projection='3d')
ax.plot_surface(X, Y, Z-Z1, rstride=4, cstride=4)
ax.view_init(70, 30)
ax.set_zlim(-zlim, zlim)

ax = fig.add_subplot(2,2,3, projection='3d')
ax.plot_surface(X, Y, Z-Z1, rstride=4, cstride=4)
ax.view_init(0,0)
ax.set_zlim(-zlim, zlim)

ax = fig.add_subplot(2,2,4, projection='3d')
ax.plot_surface(X, Y, Z-Z1, rstride=4, cstride=4)
ax.view_init(0,90)
ax.set_zlim(-zlim, zlim)

fig.tight_layout()
fig.savefig("sample_errs.pdf")
```
