```python
import numpy as np
from math import exp, log
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%pylab inline
#%matplotlib notebook
```

```python
f = open("sets.txt", "r")
lines = f.readlines()
f.close()
```

```python
xs = np.linspace(-4, 4)
def sigmoid(a, b, x):
    return 1 / (1 + exp(b*(x - a)))
```

```python
f, axss = plt.subplots(len(lines), 2, figsize=(2*6.4, len(lines)*2.8))
for i, line in enumerate(lines):
    a, b, c, d, _ = line.split(" ")
    
    axss[i, 0].plot(xs, sigmoid(float(a), float(b), xs))
    axss[i, 1].plot(xs, sigmoid(float(c), float(d), xs))
    #axss[i, 2].plot(xs, sigmoid(float(a), float(b), xs) * sigmoid(float(c), float(d), xs))
    
f.savefig("sets2.pdf")
```

```python
f = open("errs.txt", "r")
lines = f.readlines()
f.close()

iter1 = np.empty(len(lines))
errs1 = np.empty(len(lines))

for i, line in enumerate(lines):
    ii, e = line.split(" ")
    iter1[i] = ii
    errs1[i] = float(e)
```

```python
f = open("errs2.txt", "r")
lines = f.readlines()
f.close()

iter2 = np.empty(len(lines))
errs2 = np.empty(len(lines))

for i, line in enumerate(lines):
    ii, e = line.split(" ")
    iter2[i] = ii
    errs2[i] = float(e)
```

```python
figure(figsize(6.4, 4.8))
plt.plot(iter1, log(errs1), label='stohastiki')
plt.plot(iter2, log(errs2), label='pravi gradijent')
plt.legend();
plt.savefig("errs.pdf")
```

```python
figure(figsize(6.4, 4.8))
plt.plot(iter1, errs1, label='stohastiki')
plt.plot(iter2, errs2, label='pravi gradijent')
plt.legend();
plt.savefig("errs2.pdf")
```
